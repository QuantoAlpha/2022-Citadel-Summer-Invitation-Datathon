```python
def convert_emp_length(row):
    temp = row
    if temp != temp: return np.nan
    if temp[0] == '<': return 0 
    if temp[1] == '0': return 10 
    return int(temp[0])
def clean_emp_length(df_demo):
    # less than 1 treated as 0 , >10 treated as 10 
    # df_demo = df_demo.dropna()
    res = df_demo['emp_length'].apply(convert_emp_length)
    mean = res.sum()/df_demo['emp_length'].notna().sum()
    def convert_emp_length_2(row):
      temp = row
      if temp != temp: return mean 
      return row
      return int(temp[0])
    res_final = res.apply(convert_emp_length_2)
    return res_final

def convert_verification_status(row):
    temp = row
    if temp[0] == 'S': return 0 
    if temp[0] == 'N': return 1 
    if temp[0] == 'V': return 2 
def clean_verification_status(df_demo):
    # map 3 status to 0, 1, 2 
    res = df_demo['verification_status'].apply(convert_verification_status)
    return res

def convert_home_ownership(row):
    temp = row
    if temp[0] == 'M': return 0 
    if temp[0] == 'R': return 1 
    if temp[0] == 'O': return 2 
    if temp[0] == 'A': return 3 
    return np.nan
def clean_home_ownership(df_demo):
    # less than 1 treated as 0 , >10 treated as 10 
    res = df_demo['home_ownership'].apply(convert_home_ownership)
    res = res.dropna()
    return res



def clean_annual_inc(df_demo):
    # nothing change
    # df_demo = df_demo.dropna()
    res = df_demo["annual_inc"]
    mean = res.sum()/res.notna().sum()
    def convert_annual_inc_2(row):
      temp = row
      if temp != temp: return mean 
      
      return row 
    res_final = res.apply(convert_annual_inc_2)
    return res_final
# def clean_dti(df_demo):
#     # nothing changes 
#     return df_demo

df1['emp_length'] = clean_emp_length(df1)
df1['verification_status'] = clean_verification_status(df1)
df1['home_ownership'] = clean_home_ownership(df1)
df1['annual_inc'] = clean_annual_inc(df1)

```

```python
def one_hot_veri(row):
    if row == 0: 
        return pd.Series([1,0,0])
    if row == 1: 
        return pd.Series([0,1,0])
    return pd.Series([0,0,1]) 
temp1 = df1['verification_status'].apply(one_hot_veri)
# rename 
```

```python
def one_hot_home(row):
    if row == 0: 
        return pd.Series([1,0,0,0])
    if row == 1: 
        return pd.Series([0,1,0,0])
    if row == 2: 
        return pd.Series([0,0,1,0])
    return pd.Series([0,0,0,1]) 
temp2 = df1['home_ownership'].apply(one_hot_home)
```
